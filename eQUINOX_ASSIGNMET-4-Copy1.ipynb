```python
import qiskit
import numpy as np
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit,Aer, execute
from qiskit.tools.visualization import plot_histogram
from qiskit.providers.fake_provider import FakeBogotaV2 
from qiskit.circuit import Gate
from math import pi
x = 0
y = 1
sim = Aer.get_backend('aer_simulator')
backend_bogota = FakeBogotaV2()
```

```python
swap_qc = QuantumCircuit(2)

swap_qc.swap(0,1)

swap_qc.draw(output= 'mpl')
```

```python
swap_qc_cnot = QuantumCircuit(2)
swap_qc_cnot.cx(x,y) # copies 1 from x to y
swap_qc_cnot.cx(x,y) # uses the 1 on y to rotate the state of x to 0
swap_qc_cnot.draw()
```

```python
swap_qc_cnot.cx(y,x)
swap_qc_cnot.cx(x,y)
swap_qc_cnot.cx(y,x)
swap_qc_cnot.draw(output = 'mpl')
```

```python
double_swap = QuantumCircuit(2)

double_swap.h(0)

double_swap = double_swap.compose(swap_qc)
double_swap.compose(swap_qc_cnot)


double_swap.measure_all()


double_swap.draw()
```
